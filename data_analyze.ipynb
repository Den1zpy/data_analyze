```python
import pandas as pd

data = {'product_name': ['Samsung Galaxy S21', 'Apple iPhone 12 Pro', 'Xiaomi Redmi Note 10', 'Samsung 55TU8000 Smart TV', 'LG 65NANO796 Smart TV', 'Sony WH-1000XM4 Wireless Headphones', 'Apple AirPods Pro Wireless Headphones', 'Logitech MX Master 3 Wireless Mouse', 'Microsoft Surface Laptop 4', 'Lenovo Yoga C940 Laptop'],
        'category': ['Phones', 'Phones', 'Phones', 'TVs', 'TVs', 'Headphones', 'Headphones', 'Computer Accessories', 'Computers', 'Computers'],
        'price': [9999, 11999, 2999, 4999, 7999, 2499, 1999, 699, 9999, 8499],
        'star_rating': [4.5, 4.7, 4.2, 4.6, 4.5, 4.8, 4.7, 4.9, 4.6, 4.7],
        'reviews_count': [2058, 1932, 4657, 678, 423, 1123, 2345, 687, 324, 416]
        }

df = pd.DataFrame(data)
df.to_csv('urunler.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('urunler.csv')


category_counts = df['category'].value_counts()


plt.bar(category_counts.index, category_counts.values)


plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Product Counts by Category on xyz corp')


plt.show()
```

```python
import seaborn as sns


sns.boxplot(x='category', y='price', data=df)


plt.xlabel('Category')
plt.ylabel('Price')
plt.title('Product Prices by Category on XYZ corp')


plt.show()
```

```python

plt.scatter(df['price'], df['star_rating'], alpha=0.5)


plt.xlabel('Price')
plt.ylabel('Star Rating')
plt.title('Product Prices vs Star Ratings on XYZ corp')


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('urunler.csv')


fig, ax = plt.subplots(figsize=(10, 6))
df.boxplot(column='price', by='category', ax=ax)
ax.set_xlabel('Kategori')
ax.set_ylabel('Fiyat (TL)')
ax.set_title('Ürünlerin Fiyat Dağılımı')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('urunler.csv')


df['purchases'] = df['reviews_count'] / df['star_rating']


plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x='price', y='purchases', hue='category', s=100)


plt.title('Fiyat ve Satın Alınma Sayısı İlişkisi', fontsize=16)
plt.xlabel('Fiyat', fontsize=12)
plt.ylabel('Satın Alınma Sayısı', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=10)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# veri setini yükle
df = pd.read_csv('urunler.csv')

# yıldız puanına göre ürünleri sırala
df_sorted = df.sort_values(by='star_rating', ascending=False)

# en yüksek yıldız puanına sahip ilk 10 ürünü seç
top_10_products = df_sorted.head(10)

# görselleştirme yap
plt.barh(top_10_products['product_name'], top_10_products['star_rating'])
plt.xlabel('Star Rating')
plt.ylabel('Product Name')
plt.title('Top 10 Products by Star Rating')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('urunler.csv')

# Fiyatları büyükten küçüğe sırala
df = df.sort_values('price', ascending=False)

# Grafik boyutunu belirle
fig, ax = plt.subplots(figsize=(10, 6))

# Bar grafiğini çiz
ax.bar(df['product_name'], df['price'])

# X ekseni etiketlerini 90 derece çevir
plt.xticks(rotation=90)

# Eksen etiketlerini belirle
ax.set_xlabel('Ürün')
ax.set_ylabel('Fiyat')

# Grafiği göster
plt.show()
```
